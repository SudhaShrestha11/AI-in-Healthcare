```python
import pandas as pd
import numpy as np
```

```python
diagnosis = pd.read_csv('DIAGNOSES_ICD.csv').set_index('ROW_ID')
diagnosis.head()
```

```python
notes=pd.read_csv('NOTEEVENTS.csv.gz', compression='gzip', header=0, sep=',', quotechar='"').set_index('ROW_ID')
notes.head()
```

```python
diagnosis.dtypes
```

```python
notes['CATEGORY'].unique()
```

```python
notes = notes[notes['CATEGORY'].str.contains('Respiratory')]
notes.head()
```

```python
notes.shape
```

```python
merged_notes = pd.merge(diagnosis,notes[['SUBJECT_ID','HADM_ID', 'CATEGORY', 'DESCRIPTION', 'TEXT']], on=('SUBJECT_ID','HADM_ID'), how = 'inner')
merged_notes.head(25)
```

```python
new_notes = merged_notes[['SUBJECT_ID','HADM_ID', 'ICD9_CODE', 'TEXT']].reset_index(drop=True)
new_notes.head()
```

```python
filtered_notes = new_notes[new_notes['ICD9_CODE'] == '51881'].reset_index(drop=True)
filtered_notes.head(15)
```
