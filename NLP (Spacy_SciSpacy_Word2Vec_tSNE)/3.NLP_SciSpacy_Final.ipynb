```python
# import medically trained NLP package (en_ner_bc5cdr_md)
import en_ner_bc5cdr_md
nlp = en_ner_bc5cdr_md.load()
```

```python
doc1 = []
for i in range(len(text)):
  doc1.append(nlp(text[i]))
  for ent in doc1[-1].ents:
    print(ent.text, ent.start_char, ent.end_char, ent.label_)
    print('****************************************************************************************************************')
```

```python
#Entity Visualizer
from spacy import displacy
for i in range(len(doc1)):
  displacy.render(doc1[i], style="ent", jupyter=True)
  print('********************************************************************************************************************')
```
