```python
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
import pandas as pd
import numpy as np
```

```python
notes_updated = pd.read_csv('notes_updated_df.csv')
notes_updated
```

```python
notes_updated.info()
```

```python
# Tokenize the text field
doc = notes_updated['TEXT']
for token in doc:
  print(token)
  print('****************************************************************************************************************')
```

```python
#Append the text data
doc_append = []
for i in range(len(doc)):
      doc_append.append(nlp(doc[i]))
      print(doc_append[-1])
      print('****************************************************************************************************************')
```

```python
# Clean up text by removing any punctuation marks
token_wo_punc = []
for i in range(len(doc)):
  token_wo_punc.append([token.orth_ for token in nlp(doc[i]) if not token.is_punct or token.is_space])
  print(token_wo_punc[-1])
  print('****************************************************************************************************************')
```

```python
#Lemmatize
token_lemma_pos = []
for i in range(len(token_wo_punc)):
  token_lemma_pos.append([(token.text,token.lemma_, token.pos_) for token in nlp(' '.join(token_wo_punc[i]))])
  print(token_lemma_pos[-1])
  print('****************************************************************************************************************')
```

```python
#Named Entities Recognition
token_NE = []
for ent in doc:
  token_NE.append([(ent.text, ent.start_char, ent.end_char, ent.label_) for ent in nlp(ent).ents])
  print(token_NE[-1])
  print('****************************************************************************************************************')
```

```python
from spacy import displacy
for i in range(len(doc_append)):
  displacy.render(nlp(doc_append[i]), style='ent', jupyter=True)
  print('****************************************************************************************************************')
```

```python
#sentence identifier
for i in range(len(doc_append)):
  for ix, sent in enumerate(doc_append[i].sents, 1):
    print("Sentence number {}:{}".format(ix,sent))
  print('**************************************************************************************************************************')
```

```python
# Dependence tree
for i in range(len(doc_append)):
  sentence_spans = list(doc_append[i].sents)
  displacy.render(sentence_spans, style='dep', jupyter=True)
```
